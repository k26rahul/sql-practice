```python
import sqlite3
import pandas as pd
```

```python
conn = sqlite3.connect("database.sqlite")
```

```python
def run_sql(sql):
  sql = sql.strip()
  cur = conn.cursor()
  try:
    cur.execute(sql)

    if sql.lower().startswith("select"):
      rows = cur.fetchall()
      cols = [desc[0] for desc in cur.description]
      df = pd.DataFrame(rows, columns=cols)
      print(df)
    else:
      conn.commit()
      print("OK")
  except Exception as e:
    print("Error:", e)
```

```python
run_sql("""
CREATE TABLE students(
    id INTEGER PRIMARY KEY,
    name TEXT,
    marks INTEGER
);
""")
```

```python
run_sql("INSERT INTO students (name, marks) VALUES ('Paru', 55);")
```

```python
run_sql("SELECT * FROM students;")
```
